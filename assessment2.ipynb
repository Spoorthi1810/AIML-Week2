```python
import pandas as pd
import numpy as np
```

```python
# ==============SECTION A=================
data = {
    "Name": ["Amit", "Sara", "John", "Priya", "David"],
    "Age": [22, 25, 23, 24, 26],
    "Department": ["IT", "HR", "IT", "Finance", "HR"],
    "Salary": [40000, 42000, 38000, 45000, 41000]
}

df = pd.DataFrame(data)

# Q1: Display DataFrame
print("\nQ1: Display DataFrame")
print(df)

# Q2: Display first 3 rows
print("\nQ2: Display first 3 rows")
print(df.head(3))

# Q3: Display last 2 rows
print("\nQ3: Display last 2 rows")
print(df.tail(2))

# Q4: Display column names
print("\nQ4: Display column names")
print(df.columns)

# Q5: Display shape of dataset
print("\nQ5: Display shape of dataset")
print(df.shape)

# Q6: Employees with salary > 40000
print("\nQ6: Employees with salary > 40000")
print(df[df["Salary"] > 40000])

# Q7: Sort employees by salary
print("\nQ7: Sort employees by salary")
print(df.sort_values(by="Salary"))

# Q8: Add Bonus column (10% of salary)
print("\nQ8: Add Bonus column (10% of salary)")
df["Bonus"] = df["Salary"] * 0.10
print(df)

# Q9: Create BasicSalary column (Salary - Bonus)
print("\nQ9: Create BasicSalary column (Salary - Bonus)")
df["BasicSalary"] = df["Salary"] - df["Bonus"]
print(df)

# Q10: Drop Bonus column
print("\nQ10: Drop Bonus column")
df = df.drop(columns=["Bonus"])
print(df)




# ========== SECTION B ==========


print("\n========== SECTION B ==========")

# Q11: Read Student_data.csv
print("\nQ11: Read Student_data.csv")
stu = pd.read_csv("Student_data.csv")
print(stu)

# Q12: Show dataset description
print("\nQ12: Show dataset description")
print(stu.describe())

# Q13: Students scoring above 80
print("\nQ13: Students scoring above 80")
print(stu[stu["Marks"] > 80])

# Q14: Students scoring between 40 and 60
print("\nQ14: Students scoring between 40 and 60")
print(stu[(stu["Marks"] >= 40) & (stu["Marks"] <= 60)])

# Q15: Average marks per subject
print("\nQ15: Average marks per subject")
print(stu.groupby("Subject")["Marks"].mean())

# Q16: Count students in each subject
print("\nQ16: Count students in each subject")
print(stu["Subject"].value_counts())

# Q17: Subject with highest average marks
print("\nQ17: Subject with highest average marks")
print(stu.groupby("Subject")["Marks"].mean().idxmax())

# Q18: Add Pass/Fail column (Pass >= 50)
print("\nQ18: Add Pass/Fail column (Pass >= 50)")
stu["Result"] = np.where(stu["Marks"] >= 50, "Pass", "Fail")
print(stu)

# Q19: Replace missing values with mean
print("\nQ19: Replace missing values with mean")
stu.loc[0,"Marks"] = np.nan
stu["Marks"].fillna(stu["Marks"].mean(), inplace=True)
print(stu)

# Q20: Save cleaned data and read again
print("\nQ20: Save cleaned data and read again")
stu.to_csv("cleaned_students.csv", index=False)
new_df = pd.read_csv("cleaned_students.csv")
print(new_df)

```
